2.Kaggle'dan Titanic: Machine Learning from Disaster isimli yarışmanın train.csv isimli dosyasını indirin ve Pandas veri çerçevelerini kullanarak tüm veriyi yükleyin.
```python
df = pd.read_csv('train.csv')
df
```

```python
df.groupby(by='Sex').mean()['Age']
```

3.Koşullu seçim yöntemlerini de kullanarak, 30 yaşından küçük tüm yolcuların kurtulma oranını (survival ratio) hesaplayın. Daha sonra, kurtulma oranını cinsiyete göre ayrı ayrı hesaplayın.

```python
A=df.loc[lambda df: df['Age'] <30,:]
A
```

```python
A.groupby(by="Sex").mean()["Survived"]
```
